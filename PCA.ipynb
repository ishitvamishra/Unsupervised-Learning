```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.decomposition import PCA

x, y = make_blobs(n_samples=100, n_features=5, centers=3, random_state=42)

print("Original shape of the dataset", x.shape)

pca = PCA(n_components=2)

x_reduced = pca.fit_transform(x)
print("Reduced shape of the dataset", x_reduced.shape)

plt.figure(figsize=(8, 6))
plt.scatter(x_reduced[:, 0], x_reduced[:, 1], c=y)
plt.xlabel("First Principal Component")
plt.ylabel("Second Principal Component")
plt.show()


```

```python
X,y = make_blobs(n_samples=300,n_features=5,centers=3,random_state=42)
print("Original shape of two dataset:",X.shape)

pca = PCA(n_components=3)
x_reduced = pca.fit_transform(X)

print("Reduced shape of the dataset:",x_reduced.shape)

fig = plt.figure(figsize=(10,7))

ax = fig.add_subplot(111,projection='3d')
ax.scatter(x_reduced[:,0],x_reduced[:,1],x_reduced[:,2],c=y)
ax.set_xlabel("Principal Component 1")
ax.set_ylabel("Principal Component 2")
ax.set_zlabel("Principal Component 3") # Added z-label
ax.set_title("PCA of the dataset")
# plt.colorbar(label="cluster label")
plt.show()
```
